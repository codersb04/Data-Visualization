```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
```

```python
automobile_df = pd.read_csv("Automobile.csv")
```

```python
automobile_df.head()
```

```python
automobile_df.shape
```

## Data Visualization

### Univarite DIst plot

```python
# DIstribution PLot
sns.displot(automobile_df['normalized_losses'])
plt.show()
```

```python
sns.displot(automobile_df['engine_size'])
plt.show()
```

```python
sns.displot(automobile_df['city_mpg'], kde = False, rug=True)
plt.show()
```

### Bivariate Dist plot

```python
# kind = scatter, reg, resid, kde, hex
sns.jointplot(x = 'engine_size', y = 'horsepower', data = automobile_df, kind = 'kde')
plt.show()
```

### Multivariate DIst plot

```python
sns.pairplot(automobile_df[['horsepower', 'normalized_losses', 'engine_size']] )
```

### Plot between continuous and categorical values

```python
sns.stripplot(x = 'fuel_type', y = 'horsepower', data = automobile_df)
```

```python
sns.swarmplot(x = 'fuel_type', y = 'horsepower', data = automobile_df)
```

```python
sns.boxplot(x = 'number_of_doors', y = 'horsepower', data = automobile_df, hue = automobile_df['fuel_type'])
```

### Plot for amount of accurance

```python
sns.barplot(x = 'body_style', y = 'horsepower', data=automobile_df, hue = automobile_df['engine_location'])
```

```python
# Frequency of the variable
sns.countplot(x = 'body_style', data =automobile_df, hue = automobile_df['engine_location'])
```

```python
sns.pointplot(x ='fuel_system', y = 'horsepower', data=automobile_df, hue = automobile_df['number_of_doors'])
```

### Multi panel Categorical Plot

```python
# plot between multiple categorical value
sns.catplot(x= 'fuel_type',
              y ='horsepower',
              hue ='number_of_doors',
              col ="engine_location",
              data = automobile_df,
              kind="violin")

# kind:  Options are: â€œstripâ€, â€œswarmâ€, â€œboxâ€, â€œviolinâ€, â€œboxenâ€, â€œpointâ€, â€œbarâ€, or â€œcountâ€.
```

### linear Regression Plot

```python

sns.lmplot(x ='horsepower', y ='peak_rpm', data=automobile_df, hue = 'fuel_type')
```
