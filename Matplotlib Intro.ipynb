## Matlplotlib </br>
Is useful for making plots and graph for data visualization

```python
# Importing Matplotlib library
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for plot
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
# Sine Wave
plt.plot(x,y)
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show()
```

```python
# adding title x-axis and y-axis label
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Sine Value')
plt.title('Sine wave')
plt.show()
```

```python
# Parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
plt.plot(x,y,'bx')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

## Bar Plot

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50,150,40,70]
ax.bar(languages, people)
plt.xlabel('Languages')
plt.ylabel('Number of People')
plt.show()
```

## Pie Chart:</br> 
Used to know the distribution of the data in the dataset

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50,150,40,70]
ax.pie(people,labels=languages, autopct = '%1.1f%%')
plt.show()
```

# 3D scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20* np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z, c=z,cmap = 'Blues')
plt.show()
```
